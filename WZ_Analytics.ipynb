```
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.io as pio
import plotly.express as px
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime
import plotly.express as px
import dateutil.parser
```

```
def plot_xy(df, x_col, y_cols, anno=False, title='...'):

    fig = go.Figure()
    
    for y_col in y_cols:
        trace = go.Scattergl(
            x = df.datetime,
            y = df[y_col],
            mode = 'markers',
            name = y_col
        )
        fig.add_trace(trace)

    fig.update_layout(
        title=title,
        xaxis_title=x_col,
        hovermode='x')
    fig.update_yaxes(categoryorder="category ascending")
    fig.show()

def plot_xy_p(df, x_col, y_cols, anno=False, title='...'):

    fig = px.scatter(df, x=df.datetime, y='cod_tracker_match_placement', color="sbmm_match_difficulty",
                    size='cod_tracker_kills', title=title, hover_data=final_json_columns)
    fig.update_layout(yaxis=dict(type='linear',categoryorder='max descending', autorange='reversed'))
    fig.show()
```

# Load csv

```
filename_summary = './summary.csv'
```

```
df = pd.read_csv(filename_summary)
```

```
final_json_columns = ['cod_tracker_mode', 'cod_tracker_match_placement', 'cod_tracker_kills', 'cod_tracker_damage', 'sbmm_match_difficulty', 'sbmm_avg_kd', 'sbmm_median_kd']

dt = datetime.today()
#Check if there is "Today"
df.cod_tracker_date = df.cod_tracker_date.apply(lambda x: str(dt.strftime("%b"))+' '+str(dt.day) if 'Today' in x else x)
#Detect different date types
df.cod_tracker_date = df.cod_tracker_date.apply(lambda x: format(dateutil.parser.parse(x), '%b %d %Y'))

df['datetime'] = pd.to_datetime(df['cod_tracker_date'] + ' ' + df['cod_tracker_time'],  format='%b %d %Y %I:%M %p')

df['cod_tracker_match_placement'] = df['cod_tracker_match_placement'].map(lambda x: str(x)[:-2])
df.cod_tracker_match_placement = df.cod_tracker_match_placement.apply(lambda x: "0" if "n" in x else x)

df['cod_tracker_match_placement'] = pd.to_numeric(df['cod_tracker_match_placement'])
# df
```

### Plot with scatter

```
plot_xy(df, 'datetime', final_json_columns, " ")
```

### Plot with express

```
plot_xy_p(df, 'time', df.columns, ' ')
```

### Plot Parallel Categories

```
fig = px.parallel_categories(df, dimensions=None
                             ,color="cod_tracker_kills"
                             , color_continuous_scale=px.colors.sequential.Oryel
                             , dimensions_max_cardinality=40)
fig.show()
```

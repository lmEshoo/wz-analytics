```
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.io as pio
import plotly.express as px
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime
import plotly.express as px
```

```
def plot_xy(df, x_col, y_cols, anno=False, title='...'):

    fig = go.Figure()
    
    for y_col in y_cols:
        trace = go.Scattergl(
            x = df.datetime,
            y = df[y_col],
            mode = 'lines',
            name = y_col
        )
        fig.add_trace(trace)

    fig.update_layout(
        title=title,
        xaxis_title=x_col,
        hovermode='x')
    fig.update_yaxes(categoryorder="category ascending")
    fig.show()

def plot_xy_p(df, x_col, y_cols, anno=False, title='...'):

    fig = px.scatter(df, x=df.datetime, y='sbmm_match_difficulty', color="cod_tracker_match_placement",
                    size='cod_tracker_kills', title=title, hover_data=final_json_columns)
    fig.show()
```

```
df = pd.read_csv(filename_summary)
```

```
final_json_columns = ['cod_tracker_mode', 'cod_tracker_match_placement', 'cod_tracker_kills', 'cod_tracker_damage', 'sbmm_match_difficulty', 'sbmm_avg_kd', 'sbmm_median_kd']
df['datetime'] = pd.to_datetime(df['cod_tracker_date'] + ' ' + df['cod_tracker_time'],  format='%b %d %I:%M %p')
df
```

### Plot with scatter

```
plot_xy(df, 'datetime', df.columns, " ")
```
